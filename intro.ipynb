```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

### Explorer Datasets ###

```python
print("\nCampos to dataset Iris ({0}): {1}".format(iris.filename, iris.feature_names))
print("\nClasses to dataset Iris: {0}".format(iris.target_names))
print("\nPrimeiros registros do dataset Iris: {0}".format(X[:5]))
print("\nPrimeiros registros das classes: {0}".format(y[:5]))
```

### Visualização de dados ###

```python
iris_dataframe = pd.DataFrame(np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
ax2 = pd.plotting.scatter_matrix(iris_dataframe, c=iris['target'], figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
ax3 = sns.pairplot(iris_dataframe, hue='target')
```
